#### **Set 1-1**

```python
import pandas as pd
import numpy as np
import io
import requests
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
url = 'https://raw.githubusercontent.com/rgbaldov/LBYMF3B/main/company_sales.csv'
s=requests.get(url).content
df =pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
x = df.month
y = df.total_profit
plt.plot(x,y, color = '#000000')
```

#### **Set 1-2**

```python
y = df.mouse + df.keyboard + df.mousepad + df.headphone + df.speaker + df.router
plt.scatter(x,y, color = '#000000')
```

#### **Set 1-3**

```python
y = df.total_profit
plt.plot(x,y, color = '#087830', linestyle = 'dotted', linewidth = '3', marker = 's', mfc='#003a6c', 
         mec='#003a6c')
plt.ylabel('Sold Units')
plt.xlabel('Month')
plt.legend(['Sold units per month'], loc = 'lower right')
```

#### **Set 1-4**

```python
y1 = df.mouse
y2 = df.keyboard 
y3 = df.mousepad 
y4 = df.headphone 
y5 = df.speaker 
y6 = df.router

fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(x,y1, color = '#000000', marker = 'o', linewidth = '3')
ax1.plot(x,y2, color = 'blue', marker = 'o', linewidth = '3')
ax1.plot(x,y3, color = 'red', marker = 'o', linewidth = '3')
ax1.plot(x,y4, color = 'green', marker = 'o', linewidth = '3')
ax1.plot(x,y5, color = 'maroon', marker = 'o', linewidth = '3')
ax1.plot(x,y6, color = 'yellow', marker = 'o', linewidth = '3')
ax1.legend(['Mouse', 'Keyboard', 'Mousepad', 'Headphone', 'Speaker', 'Router'], loc = 'upper left')
```

#### **Set 1-5**

```python
fig, ax2 = plt.subplots(figsize=(12,8))

ax2.bar(x, y1, width = 0.4, align = 'center', color = '#087830')
ax2.bar(x, y2, width = 0.4, align = 'edge', color = '#003a6c')
plt.legend(['Mouse', 'Keyboard'], loc = 'upper left')
```

#### **Set 1-6**

```python
s1 = linregress (x,y1)
m1 = s1.slope
b1 = s1.intercept
s2 = linregress (x,y2)
m2 = s2.slope
b2 = s2.intercept
s3 = linregress (x,y3)
m3 = s3.slope
b3 = s3.intercept
s4 = linregress (x,y4)
m4 = s4.slope
b4 = s4.intercept
s5 = linregress (x,y5)
m5 = s5.slope
b5 = s5.intercept
s6 = linregress (x,y6)
m6 = s6.slope
b6 = s6.intercept

fig, ax = plt.subplots(figsize=(12,8))

ax.plot(x,y1, color = '#000000')
ax.plot(x,y2, color = 'blue')
ax.plot(x,y3, color = 'red')
ax.plot(x,y4, color = 'green')
ax.plot(x,y5, color = 'maroon')
ax.plot(x,y6, color = 'yellow')
ax.plot(x, m1*x + b1, color = 'black', linestyle = '--')
ax.plot(x, m2*x + b2, color = 'blue', linestyle = '--')
ax.plot(x, m3*x + b3, color = 'red', linestyle = '--')
ax.plot(x, m4*x + b4, color = 'green', linestyle = '--')
ax.plot(x, m5*x + b5, color = 'maroon', linestyle = '--')
ax.plot(x, m6*x + b6, color = 'yellow', linestyle = '--')
ax.legend(['Mouse', 'Keyboard', 'Mousepad', 'Headphone', 'Speaker', 'Router'], loc = 'upper left')
```

```python
print('Mouse\nSlope: {0:.2f}\nIntercept: {1:.2f}'.format(m1,b1))
print('\nKeyboard\nSlope: {0:.2f}\nIntercept: {1:.2f}'.format(m2,b2))
print('\nMousepad\nSlope: {0:.2f}\nIntercept: {1:.2f}'.format(m3,b3))
print('\nHeadphone\nSlope: {0:.2f}\nIntercept: {1:.2f}'.format(m4,b4))
print('\nSpeaker\nSlope: {0:.2f}\nIntercept: {1:.2f}'.format(m5,b5))
print('\nRouter\nSlope: {0:.2f}\nIntercept: {1:.2f}'.format(m6,b6))
```
