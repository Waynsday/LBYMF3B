```python
import pandas as pd
import numpy as np
import io
import requests
import matplotlib.pyplot as plt
from scipy.stats import linregress
import sklearn as sk
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ClassificationReport, ConfusionMatrix
import warnings
warnings.filterwarnings('ignore')
```

### **MRE1 Dataset**

```python
url = 'https://raw.githubusercontent.com/rgbaldov/LBYMF3B/main/concrete_data.csv'
s=requests.get(url).content
df =pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
y = df.compressive_strength
x1 = df.cement
x2 = df.blast_furnace_slag
x3 = df.fly_ash
x4 = df.water
x5 = df.superplasticizer
x6 = df.coarse_aggregate
x7 = df.fine_aggregate
x8 = df.age
```

#### **Compressive Strength vs Cement**

```python
s1 = linregress(x1, y)
m1 = s1.slope
b1 = s1.intercept

fig, ax = plt.subplots(figsize=(12,8))

ax.plot(x1, m1*x1 + b1, color = 'red', linestyle = '--')
ax.scatter(x1, y, color = 'black')
print('Slope: {:.2f}'.format(m1))
print('Intercept: {:.2f}'.format(b1))
print('Regression Coefficient: {:.2f}'.format(m1))
print('Mean Squared Error: {:.2f}'.format(mean_squared_error(y, m1*x1 + b1)))
print('Coefficient of Determination: {:.2f}'.format(r2_score(y, m1*x1 + b1)))
```

### **BME1 Dataset**

```python
url = 'https://raw.githubusercontent.com/rgbaldov/LBYMF3B/main/heart_failure_clinical_records.csv'
s=requests.get(url).content
df1 =pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
id = [1,1,8,2,0,9,9,3]
if sum(id)%3 == 0:
    size = 0.3
    print('Use 70:30')
else:
    size = 0.4
    print('Use 60:40')
```

```python
label_names = df1.columns.values[12]
labels = df1[label_names]
feature_names = df1.columns.values[0:12]
features = df1[feature_names]
```

```python
classes = ['Not dead', 'Dead']
train, test, train_labels, test_labels = train_test_split(features, labels, test_size = size, random_state = 42)
```

##### **Gaussian NB**

```python
gnb = GaussianNB()
model = gnb.fit(train,train_labels)
pred = gnb.predict(test)
visualizer = ClassificationReport(gnb, classes=classes)
visualizer.fit(train,train_labels)
visualizer.score(test,test_labels)
visualizer.poof()
```

```python
plt.figure()
vis = ConfusionMatrix(gnb, classes = classes, percent = True)
plt.title("Confusion Matrix", fontsize = 18)
plt.xlabel("Predicted Class", fontsize = 14)
plt.ylabel("True Class", fontsize = 14)
vis.fit(train,train_labels)
vis.score(test,test_labels)
g = vis.poof()
```

```python
gnb_accuracy = accuracy_score(test_labels, pred)
```

##### **Multinomial NB**

```python
mnb = MultinomialNB()
model1 = mnb.fit(train,train_labels)
pred1 = mnb.predict(test)
visualizer1 = ClassificationReport(mnb, classes=classes)
visualizer1.fit(train,train_labels)
visualizer1.score(test,test_labels)
visualizer1.poof()
```

```python
plt.figure()
vis1 = ConfusionMatrix(mnb, classes = classes, percent = True)
plt.title("Confusion Matrix", fontsize = 18)
plt.xlabel("Predicted Class", fontsize = 14)
plt.ylabel("True Class", fontsize = 14)
vis1.fit(train,train_labels)
vis1.score(test,test_labels)
g1 = vis1.poof()
```

```python
mnb_accuracy = accuracy_score(test_labels, pred1)
```

##### **Complement NB**

```python
cnb = ComplementNB()
model2 = cnb.fit(train,train_labels)
pred2 = cnb.predict(test)
visualizer2 = ClassificationReport(cnb, classes=classes)
visualizer2.fit(train,train_labels)
visualizer2.score(test,test_labels)
visualizer2.poof()
```

```python
plt.figure()
vis2 = ConfusionMatrix(cnb, classes = classes, percent = True)
plt.title("Confusion Matrix", fontsize = 18)
plt.xlabel("Predicted Class", fontsize = 14)
plt.ylabel("True Class", fontsize = 14)
vis2.fit(train,train_labels)
vis2.score(test,test_labels)
g2 = vis2.poof()
```

```python
cnb_accuracy = accuracy_score(test_labels, pred2)
```

##### **Bernoulli NB**

```python
bnb = BernoulliNB()
model3 = bnb.fit(train,train_labels)
pred3 = bnb.predict(test)
visualizer3 = ClassificationReport(bnb, classes=classes)
visualizer3.fit(train,train_labels)
visualizer3.score(test,test_labels)
visualizer3.poof()
```

```python
plt.figure()
vis3 = ConfusionMatrix(bnb, classes = classes, percent = True)
plt.title("Confusion Matrix", fontsize = 18)
plt.xlabel("Predicted Class", fontsize = 14)
plt.ylabel("True Class", fontsize = 14)
vis3.fit(train,train_labels)
vis3.score(test,test_labels)
g3 = vis3.poof()
```

```python
bnb_accuracy = accuracy_score(test_labels, pred3)
```

```python
print('Accuracy\nGaussianNB: {0:.2f}\nMultinomialNB: {1:.2f}\nComplementNB: {2:.2f}\nBernoulliNB: {3:.2f}'.
      format(gnb_accuracy,mnb_accuracy,cnb_accuracy,bnb_accuracy))
```
